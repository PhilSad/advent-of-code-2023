```python
input_file = "./input.txt"
with open(input_file, "r") as f:
    raw_data = f.read()
raw_data = raw_data.split("\n\n")
directions = raw_data[0]
network = raw_data[1].split("\n")
network = [x for x in network if x != ""]
network = {x.split(" = ")[0]: x.split(" = ")[1] for x in network}
network = {k: v[1:-1] for k, v in network.items()}
network = {k: v.split(", ") for k, v in network.items()}
network
```

```python
def find_nb_steps(cur_node):
    etapes = 0
    for direction in cycle(directions):
        etapes += 1
        if direction == "L":
            cur_node = network[cur_node][0]
        elif direction == "R":
            cur_node = network[cur_node][1]
        
        if cur_node[-1] == "Z":
            break
    return etapes
```

```python
starting_nodes = []
for k in network.keys():
    if k[-1] == "A":
        starting_nodes.append(k)
starting_nodes
```

```python
steps = []
for node in starting_nodes:
    steps.append(find_nb_steps(node))
steps
```

```python
import math
```

```python
math.lcm(*steps)
```

```python
cur_nodes = starting_nodes.copy()
etapes = 0
for direction in cycle(directions):
    etapes += 1
    new_nodes = []
    for cur_node in cur_nodes:
        if direction == "L":
            new_nodes.append(network[cur_node][0])
        elif direction == "R":
            new_nodes.append(network[cur_node][1])
    cur_nodes = new_nodes.copy()
            
    end_with_Z = []
    for cur_node in cur_nodes:
        end_with_Z.append(cur_node[-1] == "Z")
            
    if all(end_with_Z):
        break
etapes
```

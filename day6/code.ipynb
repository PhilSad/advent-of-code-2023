```python
from collections import namedtuple
Race = namedtuple("Race", "time distance")
```

```python
input_file = "./input.txt"
with open(input_file, "r") as f:
    raw_data = f.read().splitlines()
time_data = raw_data[0]
time_data = time_data.split(':')[1]
time_data = time_data.split(' ')
time_data = [d for d in time_data if d != '']
time_data = [int(d) for d in time_data]

distance_data = raw_data[1]
distance_data = distance_data.split(':')[1]
distance_data = distance_data.split(' ')
distance_data = [d for d in distance_data if d != '']
distance_data = [int(d) for d in distance_data]

data = list(zip(time_data, distance_data))
data = [Race(d[0], d[1]) for d in data]
data
```

```python
def get_distance(time_tot, time_hold):
    speed_per_hold = 1
    speed = speed_per_hold * time_hold
    distance = speed * (time_tot - time_hold)
    return distance
```

```python
def get_nb_beat(race):
    beat_n = 0
    for i in range(race.time + 1 ):
        distance = get_distance(race.time, i)
        if distance > race.distance:
            beat_n += 1
    return beat_n
```

```python
sol = 1
for race in data:
    sol *= get_nb_beat(race)
sol
```

```python
with open("./input.txt", "r") as f:
    lines = f.readlines()
```

```python
#split by :
games = {line.split(":")[0]: line.split(':')[1].strip() for line in lines}
games
```

```python
games_semicolon = {game_id: game.split(';') for game_id, game in games.items()}
# strip
games_semicolon = {game_id: [item.strip() for item in game] for game_id, game in games_semicolon.items()}
games_semicolon
```

```python
colon_splits = {game_id : [tirage.split(',') for tirage in tirages] for game_id, tirages in games_semicolon.items()}
colon_splits
```

```python
pas_ok_tirages = []
for game_id, tirages in colon_splits.items():
    for tirage in tirages:
        tirage = [tir.strip() for tir in tirage]
        
        balles = {item.split(' ')[1] : int(item.split(' ')[0]) for item in tirage}
        if balles.get('red',0) > 12 or balles.get('green', 0) > 13 or balles.get('blue', 0) > 14:
            print(tirage)
            pas_ok_tirages.append(game_id)
print(pas_ok_tirages)
pas_ok_tirages = [int(tirage.split(' ')[1]) for tirage in pas_ok_tirages]

            
            
```

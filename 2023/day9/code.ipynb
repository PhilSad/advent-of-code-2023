```python
input_file = "./input.txt"
with open(input_file, "r") as f:
    raw_data = f.read().splitlines()
raw_data = [x.split(" ") for x in raw_data]
data = [[int(x) for x in y] for y in raw_data]
data
```

```python
len(data)
```

```python
from itertools import pairwise
```

```python
def recursive_find(serie):
    # print("called with serie", serie)
    if all(x == 0 for x in serie):
        return serie + [0]
    
    new_serie = []
    for a, b in pairwise(serie):
        new_serie.append(b - a)
    
    next_serie = recursive_find(new_serie)
    
    new_element = next_serie[-1] + new_serie[-1]
    # print("new_element", new_element)
    new_serie.append(new_element)
    
    return new_serie

def pred_next_element(serie):
    new_serie = recursive_find(serie)
    return new_serie[-1] + serie[-1]
```

```python
somme = 0
for data_point in data:
    somme += pred_next_element(data_point)
somme
    
```

```python
input_file = "./input.txt"
with open(input_file, "r") as f:
    raw_data = f.read().splitlines()
raw_data = [x.split(" ") for x in raw_data]
data = [[int(x) for x in y] for y in raw_data]
data
```

```python
def recursive_find(serie):
    # print("called with serie", serie)
    if all(x == 0 for x in serie):
        return serie + [0]
    
    new_serie = []
    for a, b in pairwise(serie):
        new_serie.append(b - a)
    
    next_serie = recursive_find(new_serie)
    
    new_element =  new_serie[0] - next_serie[0]
    # print("new_element", new_element)
    new_serie.insert(0,new_element)
    
    return new_serie

def pred_next_element(serie):
    new_serie = recursive_find(serie)
    return   serie[0] - new_serie[0]
```

```python
pred_next_element(data[2])
```

```python
somme = 0
for data_point in data:
    somme += pred_next_element(data_point)
somme
```

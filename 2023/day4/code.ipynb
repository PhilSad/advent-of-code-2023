```python
with open('./input.txt') as f:
    data = f.read().splitlines()
data = {d.split(":")[0] : d.split(":")[1] for d in data}
data = {card: [elem for elem in data[card].split('|')] for card in data}
data = {card: [ [int(e) for e in elem.split(' ') if e != ''] for elem in data[card]] for card in data}

data2 = list(data.values())
data2 = [[elem] for elem in data2]

count = 0
while len (data2) > 0:
    print(len(data2))
    cur_cards = data2.pop(0)
    while len(cur_cards) > 0:
        count += 1
        grates, gagnants = cur_cards.pop(0)
        #print('cur_cards', [grates, gagnants])
        score_card = 0
        for grate in grates:
            if grate in gagnants:
                if score_card == 0:
                    score_card = 1
                else:
                    score_card *= 2
        #print('score_card', score_card)
        if score_card > len(data2):
            score_card = len(data2)
        for i in range(score_card):
            #print('adding', data2[score_card - i -1][0])
            data2[score_card - i -1].insert(0, data2[score_card - i -1][0])

count
```
